```python
print("------> End-to-End Encryption Simulation (XOR Logic) <--------")


import random

User1_message = int(input("Message Send By User 1 :  "))
seq_key = random.randint(1,100)
encryption_key = User1_message ^ seq_key
print("Message seen by Whatapp community :", encryption_key)

decryption = encryption_key ^ seq_key
print("Message seen by user 2 :",decryption)
```
